```python
# TRAINING OUR IRIS DATASET
```

```python
# IMPORTING  LOGISTIC REGRESSION FROM SKLEARN
from sklearn.linear_model import LogisticRegression
```

```python
# USING LOGISTIC REGRESSION TO PREDICT
LR=LogisticRegression()
```

```python
# FITTING THE x AND y VALUES
LR.fit(x,y)
```

```python
# PREDICTING THE x_TEST VALUE
LR.predict(x_test)
```

```python
# COMPARISON OF PREDICTION VALUES
print('Actual value',list(y_test))
print('Predicted value',list(LR.predict(x_test)))
```

```python
# ACCURACY OF THE PREDICTED VALUE
Accuracy=(LR.score(x_test,y_test)*100)
Accuracy
```
